```python
import pandas as pd
```

```python
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vRZwwifOWN0HvrziWvwR-DTk8snwlmFfeBu86CNF5knUIEEjYnFoQS82RJuILjWu6R1UnC9mzaTd9bR/pub?output=csv'
data = pd.read_csv(url)
data.head()
```

1. Import the data the drop the 'casual' and 'registered' columns.  These are redundant with your target, 'count'.  



```python
data = data.drop(['casual','registered'], axis=1) # this is to drop casual and registered
```

```python
data.head() # this is to check if the above columns were droped 
```

2. Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame containing the:

  1. Name of the Month
  2. Name of the Day of the Week
  3. Hour of the Day

```python
data['datetime'] = pd.to_datetime(data['datetime']) # this is to concert datetime type into datetime dtype
data.info()
```

```python
data['month (name)'] = data['datetime'].dt.month_name()
data['day of week (name)'] = data['datetime'].dt.day_name()
data['hour of day'] = data['datetime'].dt.hour.astype('object')
data.head()
```

Drop the 'datetime' and 'season' columns.  These are now redundant.



```python
data = data.drop(['datetime', 'season'], axis=1) # this is to drop datetime and season columns
```

3. The temperatures in the 'temp' and 'atemp' column are in Celsius.  Use `.apply()` to convert them to Fahrenheit.

```python
data['temp'] = data.apply(lambda x: (9/5)*x['temp']+32,axis=1)
data['atemp'] = data.apply(lambda x: (9/5)*x['atemp']+32,axis=1)
```

```python
data.head()
```

4. Create a new column, 'temp_variance' that is the difference between 'temp' and 'atemp'.  Drop the 'atemp' column.

```python
data['temp_variance'] = data['temp'] - data['atemp']
data.drop(columns='atemp', inplace=True)
```
